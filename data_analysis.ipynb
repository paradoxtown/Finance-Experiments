```python
import pandas as pd
```

```python
exp_results = pd.read_csv('./static/exp_results.csv')
```

### Sharpe Ratio

```python
exp_results.groupby(by=['model_name', 'tech_indicators']).describe()[['test_acc', 'test_auc', 'Sharpe Ratio']]
```

### AttLSTM

```python
att_lstm = exp_results[exp_results['model_name'] == 'AttLSTMClassifier']
```

```python
att_lstm.groupby(by='tech_indicators').describe()[['test_acc', 'test_auc']]
```

### LR

```python
lr = exp_results[exp_results['model_name'] == 'LogisticRegression']
```

```python
lr.groupby(by='tech_indicators').describe()[['test_acc', 'test_auc']]
```

### SVM

```python
svm = exp_results[exp_results['model_name'] == 'SVM']
```

```python
svm.groupby(by='tech_indicators').describe()[['test_acc', 'test_auc']]
```

### MLP

```python
mlp = exp_results[exp_results['model_name'] == 'MLP']
```

```python
mlp.groupby(by='tech_indicators').describe()[['test_acc', 'test_auc']]
```

### LSTM

```python
lstm = exp_results[exp_results['model_name'] == 'LSTMClassifier']
```

```python
lstm.groupby(by=['tech_indicators', 'n_epochs']).describe()[['test_acc', 'test_auc']]
```

### GRU

```python
gru = exp_results[exp_results['model_name'] == 'GRUClassifier']
```

```python
gru.groupby(by=['tech_indicators']).describe()[['test_acc', 'test_auc']]
```

## With vs No

```python
exp_results.drop(exp_results[exp_results['n_epochs'] == 2000].index).groupby(['model_name', 'tech_indicators']).mean()[['test_acc', 'test_auc', 'Sharpe Ratio', 'Sortino Ratio', 'Calmar Ratio', 'Return (Ann.) [%]']].to_excel('./static/with_vs_no.xlsx')
```
